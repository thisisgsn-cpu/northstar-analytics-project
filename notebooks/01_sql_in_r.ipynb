```r
hub_failure_analysis <- sqldf("

SELECT

    h.hub_name,
    h.zone,

    COUNT(d.delivery_id) AS total_deliveries,

    SUM(
        CASE
            WHEN d.delivery_status = 'Failed'
            THEN 1
            ELSE 0
        END
    ) AS failed_deliveries,

    SUM(
        CASE
            WHEN d.delivery_status = 'OnTime'
            THEN 1
            ELSE 0
        END
    ) AS successful_deliveries,

    ROUND(
        (
            SUM(
                CASE
                    WHEN d.delivery_status = 'Failed'
                    THEN 1
                    ELSE 0
                END
            ) * 100.0
        ) / COUNT(d.delivery_id),
        2
    ) AS failure_rate

FROM deliveries d

INNER JOIN hubs h
ON d.hub_id = h.hub_id

GROUP BY h.hub_name, h.zone

ORDER BY failure_rate DESC

")

hub_failure_analysis
```

```r
driver_route_analysis <- sqldf("

SELECT

    dr.driver_id,
    dr.base_zone,
    dr.employment_type,

    COUNT(d.delivery_id) AS total_deliveries,

    ROUND(
        AVG(d.manual_route_override_count),
        2
    ) AS average_route_overrides,

    ROUND(
        AVG(d.customer_rating_post_delivery),
        2
    ) AS average_customer_rating,

    ROUND(
        AVG(d.route_distance_km),
        2
    ) AS average_route_distance,

    ROUND(
        AVG(dr.driver_rating),
        2
    ) AS driver_performance_rating

FROM deliveries d

INNER JOIN drivers dr
ON d.driver_id = dr.driver_id

GROUP BY
    dr.driver_id,
    dr.base_zone,
    dr.employment_type

HAVING total_deliveries >= 3

ORDER BY
    average_route_overrides DESC,
    average_customer_rating ASC

")

driver_route_analysis
```

```r
customer_issue_analysis <- sqldf("

SELECT

    c.customer_id,
    c.customer_type,
    c.home_zone,

    COUNT(cp.complaint_id) AS total_complaints,

    SUM(
        CASE
            WHEN d.delivery_status = 'Failed'
            THEN 1
            ELSE 0
        END
    ) AS failed_deliveries,

    ROUND(
        SUM(cp.compensation_amount),
        2
    ) AS total_compensation,

    ROUND(
        AVG(c.loyalty_score),
        2
    ) AS average_loyalty_score,

    COUNT(DISTINCT o.order_id) AS total_orders

FROM customers c

LEFT JOIN complaints cp
ON c.customer_id = cp.customer_id

LEFT JOIN orders o
ON c.customer_id = o.customer_id

LEFT JOIN deliveries d
ON o.order_id = d.order_id

GROUP BY
    c.customer_id,
    c.customer_type,
    c.home_zone

HAVING total_complaints >= 2

ORDER BY
    total_complaints DESC,
    failed_deliveries DESC,
    total_compensation DESC

")

customer_issue_analysis
```

```r
vehicle_maintenance_analysis <- sqldf("

SELECT

    v.vehicle_id,
    v.vehicle_type,
    v.assigned_zone,
    v.maintenance_status,

    ROUND(
        AVG(v.battery_health_pct),
        2
    ) AS average_battery_health,

    ROUND(
        AVG(v.odometer_km),
        2
    ) AS average_odometer_km,

    COUNT(d.delivery_id) AS total_deliveries,

    SUM(
        CASE
            WHEN d.delivery_status = 'Failed'
            THEN 1
            ELSE 0
        END
    ) AS failed_deliveries,

    ROUND(
        AVG(d.fuel_or_charge_cost),
        2
    ) AS average_operational_cost,

    ROUND(
        AVG(d.customer_rating_post_delivery),
        2
    ) AS average_customer_rating

FROM vehicles v

LEFT JOIN deliveries d
ON v.vehicle_id = d.vehicle_id

GROUP BY
    v.vehicle_id,
    v.vehicle_type,
    v.assigned_zone,
    v.maintenance_status

HAVING total_deliveries >= 3

ORDER BY
    failed_deliveries DESC,
    average_operational_cost DESC

")

vehicle_maintenance_analysis
```

```r
app_performance_analysis <- sqldf("

SELECT

    a.device_type,
    a.zone_context,
    a.event_type,

    COUNT(a.event_id) AS total_events,

    ROUND(
        AVG(a.api_latency_ms),
        2
    ) AS average_api_latency,

    SUM(
        CASE
            WHEN a.success_flag = 1
            THEN 1
            ELSE 0
        END
    ) AS successful_events,

    SUM(
        CASE
            WHEN a.success_flag = 0
            THEN 1
            ELSE 0
        END
    ) AS failed_events,

    ROUND(
        AVG(c.app_engagement_score),
        2
    ) AS average_engagement_score

FROM app_events a

LEFT JOIN customers c
ON a.customer_id = c.customer_id

GROUP BY
    a.device_type,
    a.zone_context,
    a.event_type

HAVING total_events >= 5

ORDER BY
    average_api_latency DESC,
    failed_events DESC

")

app_performance_analysis
```

```r
optimized_delivery_query <- sqldf("

SELECT

    delivery_status,

    COUNT(delivery_id) AS total_deliveries,

    ROUND(
        AVG(route_distance_km),
        2
    ) AS average_route_distance,

    ROUND(
        AVG(fuel_or_charge_cost),
        2
    ) AS average_operational_cost

FROM deliveries

WHERE route_distance_km > 5

GROUP BY delivery_status

ORDER BY total_deliveries DESC

")

optimized_delivery_query
```
